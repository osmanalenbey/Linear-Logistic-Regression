# Linear Regression

Linear regression can be considered the simplest of the machine learning algorithms, yet it is still one of the most commonly used ones due to this simplicity. It is basically the attempt to fit the best linear (or affine for high dimension cases) function to the given data. If the data itself is also linear in nature, then this gives an easy but accurate result. The linear regression can be solved using two different methods:

* Ordinary least squares (OLS) (direct method)
* Gradient descent

OLS is basically the direct solution to the problem, which given the data gives the best linear fit minimizing the mean square error. While it is tempting to use a direct method, there are some shortcomings of this in real life. Real data is noisy and frequently outliers are also present in the data. OLS is sensitive to outliers in particular and can give suboptimal results for the generalization. Gradient descent is much better at rejecting outliers and also less sensitive to numerical errors that can be observed with OLS. Moreover, while the OLS has predefined number of operations which can be considered relatively complex, gradient descent complexity is adjustable based on the iteration number and it is highly likely that it can converge in small number of iterations such that complexity is much lower than OLS.

Given the data vector $\mathbf{x}$, the linear regression output is;

$y = \mathbf{x}^T \mathbf{w} + w_0$

where $\mathbf{w}$ and $w_0$ are model parameters to be found via training.

For normal equations, concatenating different data vectors as $\mathbf{X}=[\mathbf{x}_1, \mathbf{x}_2 \dots \mathbf{x}_n]^T$, this linear regression equation can be rewritten as a vector;

$\mathbf{y} = \mathbf{X} \mathbf{w} + w_0$

The goal of OLS is minimizing the squared error $(\mathbf{y} - (\mathbf{X} \mathbf{w} + w_0))^2$

For notational simplicity, we can define $\mathbf{X}\_ =[\mathbf{x}\_ _1, \mathbf{x}\_ _2 \dots \mathbf{x}\_ _n]^T$ where $\mathbf{x}\_ _i = [1, \, \mathbf{x}_i^T]^T $ and $\mathbf{w}\_ = [w_0, \mathbf{w}^T]^T$, then the squared error can be rewritten as $J(\mathbf{w}\_)=(\mathbf{y} - \mathbf{X}\_ \mathbf{w}\_)^2$

The gradient w.r.t. $\mathbf{w}\_$ is

<!-- $\nabla_{\mathbf{w}\_}J(\mathbf{w}\_) = -2 \, (\mathbf{y} - \mathbf{X}\_ \mathbf{w}\_)^T \, \mathbf{X}\_ = -2 (\mathbf{y}^T \mathbf{X}\_ - \mathbf{w}\_^T \mathbf{X}\_ ^T \mathbf{X}\_ )=0 ⇒ $ -->

$\nabla_{\mathbf{w}\_}J(\mathbf{w}\_) = -2 \, \mathbf{X}\_^T (\mathbf{y} - \mathbf{X}\_ \mathbf{w}\_) = -2 (\mathbf{X}\_^T\mathbf{y} - \mathbf{X}\_ ^T \mathbf{X}\_ \mathbf{w}\_)=0 ⇒ \mathbf{w}\_ ^{\ast} = (\mathbf{X}\_ ^T \mathbf{X}\_)^{-1} \mathbf{X}\_^T\mathbf{y} $

If we go through the complexity calculation in general the matrix multiplication $AB$ for $A$ with size $a \times b$ and $B$ with size $b \times c$ has complexity $O(abc)$. Then if the feature length of vectors $\mathbf{x}_i$ are $k$;

* The matrix multiplication $\mathbf{X}\_^T\mathbf{y}$ has complexity $O(kn)$
* The matrix multiplication $\mathbf{X}\_ ^T \mathbf{X}\_$ has complexity $O(k^2n)$
* The inverse operation has complexity $O(k^3)$
* The final multiplication of $(\mathbf{X}\_ ^T \mathbf{X}\_)^{-1}$ and $\mathbf{X}\_^T\mathbf{y}$ has complexity $O(k^2)$

Hence the overall training complexity is $O(k^3+k^2+k^2n+kn)=O(k^3+k^2n)$

For the gradient descent case the update for the $(i+1)^{\text{th}}$ iteration is;

$\mathbf{w}\_ ^{(i+1)}=\mathbf{w}\_ ^{(i)}-\alpha \, \nabla_{\mathbf{w}\_}J(\mathbf{w}\_ ^{(i)})$

Here the constant 2 in the cost function can be embedded into learning rate $\alpha$ but keeping minus sign in front is important which after distributing into parantheses becomes

$\mathbf{w}\_ ^{(i+1)}=\mathbf{w}\_ ^{(i)}-\alpha \,  \, \mathbf{X}\_^T (\mathbf{X}\_ \mathbf{w}\_^{(i)} - \mathbf{y})$

This was the update for the single iteration. The complexity is;

* The matrix multiplication $\mathbf{X}\_ \mathbf{w}\_^{(i)}$ has complexity $O(kn)$
* The subtraction operation has complexity $O(n)$
* The final matrix multiplication of $\mathbf{X}\_^T$ and $\mathbf{X}\_ \mathbf{w}\_^{(i)} - \mathbf{y}$ has complexity $O(kn)$
* The final subtraction for the update has complexity $O(k)$

Hence the overall complexity for single iteration is $O(kn+kn+k+n)=O(kn)$ and for the whole training it will be $O(n_{\text{iter}}kn)$.

Both of the complexities were given for the training part. At the inference for single sample there is a single matrix multiplication $\mathbf{x} \_^T \mathbf{w}\_$ which has $k$ multiplications and the complexity is $O(k)$ .

```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm

def mean_square_loss(X, y, w):
    m = len(X)
    return (1/m) * np.sum(np.square(X @ w - y))

# Function to generate synthetic data
def generate_data(num_samples=100, num_features=1, noise_level=2):
    X = np.random.rand(num_samples, num_features) * 10  # Features between 0 and 10
    true_w = np.random.randn(num_features + 1, 1) * 5  # Random coefficients

    # Add np.ones as the offset term
    y = np.concatenate([np.ones((num_samples, 1)), X],axis=1) @ true_w + np.random.randn(num_samples, 1) * noise_level
    return X, y, true_w

# Normal Equation Method
def normal_equation(X, y):
    X_b = np.concatenate([np.ones((X.shape[0], 1)), X],axis=1)  # Add bias term
    return np.linalg.pinv(X_b.T @ (X_b)) @ (X_b.T) @ (y)

# Gradient Descent Method
def gradient_descent(X, y, alpha=0.01, iterations=1000):
    X_b = np.concatenate([np.ones((X.shape[0], 1)), X], axis=1)
    w = np.random.randn(X_b.shape[1], 1)  # Random initialization
    m = len(X)
    for i in tqdm(range(iterations)):
        gradients = (1/m) * X_b.T @ (X_b @ w - y)
        w -= alpha * gradients
        loss = mean_square_loss(X_b, y, w)
        if i%100==0:
            print(f"Iteration: ({i+1}/{iterations}), Loss: {loss:.4f}")
    print('\n')
    return w

np.random.seed(0)

# Generate 1D Data
X_1d, y_1d, true_w_1d = generate_data(num_samples=100, num_features=1)

# Solve for 1D case
print(f"Solving for 1D case:")

print(f"Exact solution started.")
w_exact_1d = normal_equation(X_1d, y_1d)
print(f"Exact solution ended.")
print('\n')

print(f"Gradient descent solution started.")
w_gd_1d = gradient_descent(X_1d, y_1d)
print(f"Gradient descent solution ended.")
print('\n')

# Plot 1D results
plt.figure(figsize=(5, 3))
plt.scatter(X_1d, y_1d, label="Data", alpha=0.6)
plt.plot(X_1d, np.c_[np.ones((X_1d.shape[0], 1)), X_1d].dot(w_exact_1d), label="Exact Solution", color="red")
plt.plot(X_1d, np.c_[np.ones((X_1d.shape[0], 1)), X_1d].dot(w_gd_1d), label="Gradient Descent", color="green", linestyle="dashed")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.title("Linear Regression: 1D Case")
plt.show()

# Generate Multi-Dimensional Data
X_multi, y_multi, true_w_multi = generate_data(num_samples=100, num_features=3)

# Solve for Multi-D case
print(f"Solving for multi-D case:")

print(f"Exact solution started.")
w_exact_multi = normal_equation(X_multi, y_multi)
print(f"Exact solution ended.")
print('\n')

print(f"Gradient descent solution started.")
w_gd_multi = gradient_descent(X_multi, y_multi)
print(f"Gradient descent solution ended.")
print('\n')



# Print results
print("True w (1D):", true_w_1d.ravel())
print("Exact Solution w (1D):", w_exact_1d.ravel())
print("Gradient Descent w (1D):", w_gd_1d.ravel())
print('\n')

print("True w (Multi-D):", true_w_multi.ravel())
print("Exact Solution w (Multi-D):", w_exact_multi.ravel())
print("Gradient Descent w (Multi-D):", w_gd_multi.ravel())
```

# Logistic Regression

Logistic regression can be thought as the discrete counterpart of the linear regression where the outputs are class labels instead of real values. It is basically linear regression formula encapsulated in the nonlinear sigmoid function which is defined as,

$$\sigma (x) = \frac{1}{1+\exp(-x)}$$

with derivative

$$\frac{d}{dx}\sigma (x) = \frac{\exp(-x)}{(1+\exp(-x))^2}=\sigma (x)(1-\sigma (x))$$

where the easy implementation of derivative is also one reason of choice for the sigmoid as nonlinearity.

The plot of the sigmoid function and its derivative is given in the below cell. It can be interpreted as a smooth version of a step function, thus instead of a sharp transition from $0$ to $1$ a softer one is applied. This allows for well behaving derivatives and also the data falling into this transition zone can be considered as harder to be separated and is better to map them to intermediate values.

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(-10,10,100)

sigmoid = 1/(1+np.exp(-x))
sigmoid_der = sigmoid*(1-sigmoid)

# Create a grid of 2 plots next to each other
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

# Plot sigmoid function
axes[0].plot(x, sigmoid, label='Sigmoid', color='blue')
axes[0].set_title("Sigmoid Function")
axes[0].set_xlabel("x")
axes[0].set_ylabel("sigmoid(x)")
axes[0].set_ylim([-0.1, 1.1])
axes[0].hlines(xmin=-10, xmax=10, y=1, linestyle='--', color='k', label='y=1');
axes[0].legend()
axes[0].grid(True)

# Plot sigmoid derivative
axes[1].plot(x, sigmoid_der, label='Sigmoid Derivative', color='red')
axes[1].set_title("Sigmoid Derivative")
axes[1].set_xlabel("x")
axes[1].set_ylabel("sigmoid\'(x)")
axes[1].set_ylim([-0.1, 1.1])
axes[1].hlines(xmin=-10, xmax=10, y=0.25, linestyle='--', color='k', label='y=0.25');
axes[1].legend()
axes[1].grid(True)

# Show plots
plt.tight_layout()
plt.show()
```

The predicted values in the binary logistic regression case is;

$$\hat{y} = \sigma( \mathbf{x}^T \mathbf{w} + w_0)$$

We can use the same underscore notation as in linear regression to include bias term and rewrite this as $\hat{y} = \sigma( \mathbf{x}\_^T \mathbf{w}\_)$

Since the outputs of the logistic regression are class labels instead of real values using mean square error is not the best choice here. Instead binary cross entropy function is used. This way when the predicted values match the labels zero penalty is given and when there is a mismatch a positive value is obtained.

Binary cross entropy loss is;

$$\text{BCE}=-\sum_{i=1}^n (y_i \log(\hat{y}_i) + (1-y_i) \log(1-\hat{y}_i))$$

where $y_i$ is the ground truth class label and $\hat{y_i}$ is the logit predicted by the model. This can be depicted as an error function $J(\mathbf{w}\_)$ which is a function of $\mathbf{w}\_$ because $\hat{y}_i$'s depend on $\mathbf{w}\_$ . Now we need $\nabla_{\mathbf{w}\_}J(\mathbf{w}\_)$ for the gradient descent equation. In this case it is equal to

\begin{align}
\nabla_{\mathbf{w}\_}J(\mathbf{w}\_) &= \sum_{i=1}^n \frac{\partial J}{\partial \hat{y}_i} \frac{\partial \hat{y}_i}{\partial \mathbf{w}\_} \\
&= \sum_{i=1}^n \left( \frac{1-y_i}{1-\hat{y}_i} - \frac{y_i}{\hat{y}_i} \right) \underbrace{\sigma( \mathbf{x}\_ _i^T \mathbf{w}\_)}_{\hat{y}_i} (1-\underbrace{\sigma( \mathbf{x}\_ _i^T \mathbf{w}\_)}_{\hat{y}_i}) \mathbf{x}\_ _i \\
&= \sum_{i=1}^n \left( \hat{y}_i - y_i \right)  \mathbf{x}\_ _i
\end{align}

For the multiclass case, this becomes

$$\text{CE}=-\sum_{i=1}^n \sum_{j=1}^k y_{ij} \log(\hat{y}_{ij})$$

which is called categorical cross entropy loss; and where $n$ is the number of samples and $k$ is the number of classes. The class labels are assumed to be one-hot encoded vectors in the multiclass case instead of scalar values as in the binary case.

The categorical loss function necessitates a little more examination than the binary case to be understood. In binary case if the labels $y_i$ and predictions $\hat{y}_i$ are equal, the term in summation is $0$ and it is $∞$ in the worst case where $\hat{y}_i=1$ when $y_i=0$ and vice versa. For the multiclass case, when $\hat{y}_{ij}=y_{ij}$, the innermost product is $0$ as expected. For the case $\{\hat{y}_{ij}=0, \, y_{ij}=1\}$, it is also $∞$ but when $\{\hat{y}_{ij}=1, \, y_{ij}=0\}$ it is again $0$ instead of $\infty$ which can be seen as a limitation at first glance. However, since $y_i$'s are one-hot encoded vectors and $\hat{y}_{i}$'s are softmax outputs (i.e. summing to 1) when $\{\hat{y}_{ij}=1, \, y_{ij}=0\}$ case is observed there is inevitably a case of $\{\hat{y}_{ij}=0, \, y_{ij}=1\}$ as well which will make the overall inner sum $\infty$ which is the desired result.

The gradient for the multi-class loss function has the same form with the binary one

$$
\nabla_{\mathbf{w}\_}J(\mathbf{w}\_) = \sum_{i=1}^n \left( \hat{y}_i - y_i \right)  \mathbf{x}\_ _i
$$

But the difference is that $y_i$ and $\hat{y}_i$ are $c$-dimensional vectors now instead of scalars where $c$ is the number of classes.

The approach introduced above for the multi-class case is called multinomial. There is also another approach for multi-class case called One-vs-Rest (OvR) which creates $c$ different binary classifiers for $c$ classes, each classifier labeling one particular class as $1$ and all the others as $0$.

The complexity of binary logistic regression is the same with linear regression with gradient descent case because other than the exertion of sigmoid, the general training logic is the same. Thus the training complexity is $O(n_{\text{iter}}kn)$ for the binary case and for multiclass case with $c$ classes it is $O(n_{\text{iter}}knc)$ which is same for multinomial and OvR cases. For inference, the complexities are $O(k)$ and $O(kc)$.

```python
import numpy as np
import matplotlib.pyplot as plt

def generate_linearly_separable_data(num_samples=100, num_features=2, num_classes=2):
    """
    Generates a dataset with explicitly defined linear decision boundaries.
    Ensures perfect class balance and avoids undefined regions.

    Supports both binary and multiclass classification.
    """
    assert num_classes in [2, 3], "Currently, only 2-class (binary) or 3-class (multiclass) cases are supported."

    X = np.random.uniform(-10, 10, (num_samples, num_features))  # Uniformly distributed features
    y = np.zeros(num_samples, dtype=int)  # Labels initialized

    if num_classes == 2:
        # Define a random linear decision boundary: w1*x1 + w2*x2 + b = 0
        w = np.random.randn(num_features)
        b = np.random.randn()

        # Assign labels based on which side of the line a point falls
        y = (X @ w + b > 0).astype(int)  # Class 1 if positive, otherwise class 0

        params=[w,b]

    elif num_classes == 3:
        # Define two separating hyperplanes (lines in 2D)
        w1, w2 = np.random.randn(num_features), np.random.randn(num_features)
        b1, b2 = np.random.randn(), np.random.randn()

        # Create decision regions based on these hyperplanes
        region_1 = X @ w1 + b1 > 0
        region_2 = X @ w2 + b2 > 0

        # Assign class labels based on regions
        y[region_1 & region_2] = 0  # First class
        y[region_1 & ~region_2] = 1  # Second class
        y[~region_1] = 2  # Third class

        params=[w1,b1,w2,b2]

    return X, y.reshape(-1, 1) , params  # Reshape y for compatibility

def plot_linearly_separable_data(X, y, params, num_classes, title=None):

    # Plot the generated data (Only for 2D case)
    plt.figure(figsize=(5, 5))
    plt.scatter(X[:, 0], X[:, 1], c=y.ravel(), cmap='viridis', alpha=0.7)
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")

    if num_classes==2:

        # Plot the decision boundaries
        w1, b1 = params
        x_vals1 = np.linspace(X[:, 0].min(), X[:, 0].max(), 100)
        y_vals1 = - (w1[0] / w1[1]) * x_vals1 - (b1 / w1[1])

        plt.plot(x_vals1, y_vals1, 'k--', linewidth=1, label=f"{w1[0]:.2f} x1+{w1[1]:.2f} x2=0")

    elif num_classes==3:

        # Plot the decision boundaries
        w1, b1, w2, b2 = params
        x_vals1 = np.linspace(X[:, 0].min(), X[:, 0].max(), 100)
        x_vals2 = np.linspace(X[:, 0].min(), X[:, 0].max(), 100)
        # Two decision boundaries
        y_vals1 = - (w1[0] / w1[1]) * x_vals1 - (b1 / w1[1])
        y_vals2 = - (w2[0] / w2[1]) * x_vals2 - (b2 / w2[1])

        # For 2nd decision boundary take only the coordinates that it acts as a boundary
        valid_mask = (np.column_stack((x_vals2, y_vals2)) @ w1 + b1) > 0
        x_vals2, y_vals2 = x_vals2[valid_mask], y_vals2[valid_mask]

        plt.plot(x_vals1, y_vals1, 'k--', linewidth=1, label=f"{w1[0]:.2f} x1+{w1[1]:.2f} x2=0")
        plt.plot(x_vals2, y_vals2, 'k--', linewidth=1, label=f"{w2[0]:.2f} x1+{w2[1]:.2f} x2=0")

    if title:
        plt.title(title)
    plt.ylim([X[:, 1].min(), X[:, 1].max()])
    plt.legend()
    plt.show()

# Example usage
X, y, params = generate_linearly_separable_data(num_samples=300, num_classes=3)

# Plot the generated data (Only for 2D case)
plot_linearly_separable_data(X, y, params, num_classes=3, title="Linearly Separable Data")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm

# Logistic Regression Functions
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def binary_cross_entropy_loss(y_true, y_pred, epsilon=1e-9):
    m = len(y_true)
    y_pred = np.clip(y_pred, epsilon, 1 - epsilon)  # Clip predictions to prevent instability
    return - (1/m) * np.sum(y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred))


# Multiclass Logistic Regression (Softmax)
def softmax(z):
    exp_z = np.exp(z - np.max(z, axis=1, keepdims=True))
    return exp_z / np.sum(exp_z, axis=1, keepdims=True)

def cross_entropy_loss(y_true, y_pred, epsilon=1e-9):
    m = y_true.shape[0]
    y_pred = np.clip(y_pred, epsilon, 1 - epsilon)  # Clip predictions to prevent instability
    return - (1/m) * np.sum(y_true * np.log(y_pred))

def logistic_regression(X, y, num_classes=3, alpha=0.01, iterations=1000, alg_type=None):
    X_b = np.concatenate([np.ones((X.shape[0], 1)), X], axis=1)
    if alg_type=='binary':
        w = np.random.randn(X_b.shape[1], 1)  # Random initialization
    else:
        w = np.random.randn(X_b.shape[1], num_classes)  # Random initialization
        y_one_hot = np.eye(num_classes)[y.astype(int).flatten()]

    m = len(X)

    for i in tqdm(range(iterations)):
        if alg_type=='binary':
            y_pred = sigmoid(X_b @ w)
            gradients = (1/m) * X_b.T @ (y_pred - y)
        else:
            y_pred = softmax(X_b @ w)
            gradients = (1/m) * X_b.T @ (y_pred - y_one_hot)

        w -= alpha * gradients

        if alg_type=='binary':
            loss = binary_cross_entropy_loss(y, y_pred)
        else:
            loss = cross_entropy_loss(y_one_hot, y_pred)
        if i%100==0:
            print(f"Iteration: ({i+1}/{iterations}), Loss: {loss:.4f}")
    print('\n')

    if alg_type=='binary':
        return (y_pred>0.5).astype('int'), w
    else:
        return np.argmax(y_pred,axis=1), w

np.random.seed(0)

# Generate binary data
X_bin, y_bin, params_bin = generate_linearly_separable_data(num_samples=300, num_features=2, num_classes=2)

# Solve for binary case
print(f"Solving for binary case:")

print(f"Logistic regression with gradient descent solution started.")
# y_pred_bin, w_bin = binary_logistic_regression(X_bin, y_bin)
y_pred_bin, w_bin = logistic_regression(X_bin, y_bin, alg_type='binary')
accuracy=np.sum(y_bin==y_pred_bin)/len(y_bin)
print(f"Accuracy for binary case is {accuracy}")
print(f"Logistic regression with gradient descent solution ended.")
print('\n')

# Plot the generated data (Only for 2D case)
plot_linearly_separable_data(X_bin, y_bin, params_bin, num_classes=2)



# Generate multi-class data
X_multi, y_multi, params_multi = generate_linearly_separable_data(num_samples=300, num_features=2, num_classes=3)

# Solve for multi-class case
print(f"Solving for multi-class case:")

print(f"Logistic regression with gradient descent solution started.")
y_pred_multi, w_multi = logistic_regression(X_multi, y_multi, num_classes=3)
accuracy=np.sum(np.squeeze(y_multi)==y_pred_multi)/len(y_multi)
print(f"Accuracy for multi-class case is {accuracy}")
print(f"Logistic regression with gradient descent solution ended.")
print('\n')

# Plot the generated data (Only for 2D case)
plot_linearly_separable_data(X_multi, y_multi, params_multi, num_classes=3)
```
